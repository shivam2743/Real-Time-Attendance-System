```python
import pandas as pd
import cv2
import urllib.request
import numpy as np
import os
from datetime import datetime
import face_recognition

# Function to load images from a directory
def load_images(path):
    images = []
    classNames = []
    myList = os.listdir(path)
    for cl in myList:
        curImg = cv2.imread(f'{path}/{cl}')
        images.append(curImg)
        classNames.append(os.path.splitext(cl)[0])
    return images, classNames

# Function to find face encodings for a list of images
def find_encodings(images):
    encodeList = []
    for img in images:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encode = face_recognition.face_encodings(img)[0]
        encodeList.append(encode)
    return encodeList

# Function to mark attendance in a CSV file
def mark_attendance(name):
    with open("Attendance.csv", 'a+') as f:
        myDataList = f.readlines()
        nameList = [line.split(',')[0] for line in myDataList]
        if name not in nameList:
            now = datetime.now()
            dtString = now.strftime('%H:%M:%S')
            f.writelines(f'\n{name},{dtString}')

# Function to display text on the webcam image
def display_text(img, text, position, font_scale, color, thickness):
    cv2.putText(img, text, position, cv2.FONT_HERSHEY_COMPLEX, font_scale, color, thickness)

# Main function
def main():
    path = r'C:\ATTENDANCE\image_folder'
    url = 'http://192.168.137.21/cam-hi.jpg'
    # Check if the Attendance.csv file exists
    if 'Attendance.csv' in os.listdir(os.path.join(os.getcwd(), r'C:\Users\shivam\Attendance')):
        print("Attendance.csv exists.")
    else:
        df = pd.DataFrame()
        df.to_csv("Attendance.csv")

    images, classNames = load_images(path)
    encodeListKnown = find_encodings(images)
    print('Encoding Complete')

    while True:
        img_resp = urllib.request.urlopen(url)
        imgnp = np.array(bytearray(img_resp.read()), dtype=np.uint8)
        img = cv2.imdecode(imgnp, -1)
        imgS = cv2.resize(img, (0, 0), None, 0.25, 0.25)
        imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)

        facesCurFrame = face_recognition.face_locations(imgS)
        encodesCurFrame = face_recognition.face_encodings(imgS, facesCurFrame)

        for encodeFace, faceLoc in zip(encodesCurFrame, facesCurFrame):
            matches = face_recognition.compare_faces(encodeListKnown, encodeFace)
            faceDis = face_recognition.face_distance(encodeListKnown, encodeFace)
            matchIndex = np.argmin(faceDis)

            if matches[matchIndex]:
                name = classNames[matchIndex].upper()
                y1, x2, y2, x1 = faceLoc
                y1, x2, y2, x1 = y1 * 4, x2 * 4, y2 * 4, x1 * 4
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.rectangle(img, (x1, y2 - 35), (x2, y2), (0, 255, 0), cv2.FILLED)
                cv2.putText(img, name, (x1 + 6, y2 - 6), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
                mark_attendance(name)
            else:
                # Display "Face not registered" message
                display_text(img, "Face not registered", (50, 50), 1, (0, 0, 255), 2)

        cv2.imshow('Webcam', img)
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            break

    cv2.destroyAllWindows()

#if _name_ == "_main_":
 #   main()
```
